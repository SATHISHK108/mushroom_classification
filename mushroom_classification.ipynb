```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mushrooms.csv')
```

```python
df['class'] = df['class'].replace(['p','e'],['0','1'])
```

```python
x = df.drop('class', axis=1)
```

```python
col = x.columns
```

```python
x.head()
```

```python
for f in col:
    print(f ,df[f].unique(),len(df[f].unique()))
```

```python
y = df['class']
```

```python
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoding = OneHotEncoder()
X = encoding.fit_transform(x)
```

```python
X
```

```python
#x = pd.get_dummies(x,columns=col,drop_first=True)
```

```python
#x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)

clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
prediction = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, prediction, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
disp.plot()
plt.show()
```

```python
import pickle
encoding = OneHotEncoder()
X = encoding.fit_transform(x)
encoder = 'mushroom_classify_encoder.pkl'
pickle.dump(encoding,open(encoder,'wb'))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)

clf = LogisticRegression(random_state=0).fit(X_train, y_train)

model_name = 'mushroom_model.pkl'
pickle.dump(clf,open(model_name,'wb'))
```
